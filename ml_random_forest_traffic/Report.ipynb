## Random Forest and Decision Tree comparative runs

```python
forestSizes = [5, 10, 20, 30]
tree_depths = [5, 10, 15, 20, 25, 30]
results = compareModels('../data/US_Accidents_Dec19_LITE_10000.csv', sampleSize = 5000, forest_sizes = forestSizes, tree_depths = tree_depths)
```

### Get maximum accuracy parameters for each model

```python

models = list(results['Model'].unique())
print(models)

#frames = [df_1, df_2]

#concatenate dataframes
#df = pd.concat(frames, sort=False)
df_all = pd.DataFrame()

for m in models:
    rf_res = results[results['Model']==m]
    
    best = rf_res[(rf_res['Accuracy']==rf_res['Accuracy'].max())]
   
    df_all = df_all.append(best  , ignore_index=True ) ##, sort=False)
    
df_all
    
```

### Plot results

```python
groups = results.groupby(['Model'])
#for g in groups:


for g in (g for g in groups if g[0] in ["RF_SKL"]):
    modelStr = g[0]
    df = g[1][['Max Depth', 'Accuracy', 'Forest Size']]
    
    fig, axarr= plt.subplots(1, 1, figsize=(6, 4))

    axarr.set_title(f"{modelStr} Accuracy vs Depth for Different forest sizes")
        
    for forestSize in forestSizes:
        chartData = df[df['Forest Size'] == forestSize]
        axarr.plot(chartData[['Max Depth']], chartData[['Accuracy']])
        axarr.set_xlabel("Tree Depth")
        axarr.set_ylabel("Accuracy")
       
    
    #if modelStr not in ["DT","DT_SKL"]:
    legendLabels = [f"Forest Size: {s}" for s in forestSizes]
    axarr.legend(legendLabels, loc="lower right")
    plt.show()
    
    
for g in (g for g in groups if g[0] in ["DT_SKL"]):
    modelStr = g[0]
    df = g[1][['Max Depth', 'Accuracy']]
    
    
    fig, ax= plt.subplots(1, 1, figsize=(6, 4))
    ax.set_title(f"{modelStr} Accuracy vs Depth")
    ax.plot(df[['Max Depth']], df[['Accuracy']])
    ax.set_xlabel("Tree Depth")
    ax.set_ylabel("Accuracy")
    
    plt.show()
```

#### Make a test prediction

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
def random_forest_sklearn(fs, dpt, X, y):
    
    # Initializing a random forest - use same accuracy measurement as our model
    randomForestskl = RandomForestClassifier(n_estimators = fs, max_depth = dpt, bootstrap=True, criterion='entropy', oob_score = True, random_state=0)

    rows = X.shape[0]
    ycol = y.reshape((rows,))
    
    # Building trees in the forest
    print("fitting the forest")
    randomForestskl.fit(X,ycol)

    print("done fitting")
    
    # Prdict labels
    print("predicting")
    y_predicted = randomForestskl.predict(X)

    # Comparing predicted and true labels
    results = [prediction == truth for prediction, truth in zip(y_predicted, y)]

    # Accuracy
    accuracy = float(results.count(True)) / float(len(results))

    sc = randomForestskl.score(X, y, sample_weight=None)
    oob = randomForestskl.oob_score
    
    ## accuracy and score
    print("accuracy: %.4f" % accuracy)
    print("score", sc)
    #print("oob_score", oob)
    
    return accuracy, randomForestskl

from data_process import loadAndProcessData

filePath = '../data/US_Accidents_Dec19_LITE_10000.csv'
print("reading the data")
data, cols  = loadAndProcessData(filePath, 5000, 0)
colnames = data.columns[1:]

XX = data.to_numpy()

## label on first column 
y = XX[:,:1]  
X = XX[:,1:] 

acc, model = random_forest_sklearn(30, 30, X, y)
```
